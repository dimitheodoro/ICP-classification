```python
import matplotlib.pyplot as plt

categories = ['Neuroimaging', 'Ophthalmic', 'Audiological', 'Cerebral fluid dynamics', 'Electrophysiologic', 'Other miscellaneous']
papers = [31, 83, 23, 47, 15, 20]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#ffb3e6','#c2c2f0']

fig, ax = plt.subplots(figsize=(8, 8))

ax.pie(papers, labels=categories, autopct='%1.1f%%', startangle=90, shadow=True, colors=colors)

ax.axis('equal')

plt.title('Distribution of Noninvasive ICP Methods', fontsize=20)

plt.show()
```
