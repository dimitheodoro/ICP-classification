```python
import pandas as pd
import re

df_GSC = pd.read_excel('Excel files\episode GCS.xlsx')
episoodes_GSC = df_GSC['episode'].values
df_GSC
```

```python
PINAX = pd.read_excel('Excel files\episode-names.xlsx')
episode_PINAX =  PINAX['episode'].values
PINAX['full_name'] = PINAX['LNAME'].str.cat(PINAX['fname'], sep=' ')
PINAX
```

```python
j=0
COMMON_NAMES =[]
for i in episoodes_GSC:
    if i in episode_PINAX:
        COMMON_NAMES.append((PINAX[PINAX['episode']==i]['full_name'].values[0],i))
        j+=1
print(j)
COMMON_NAMES
```

```python
df_COMMON_NAMES =pd.DataFrame(COMMON_NAMES, columns=['Name', 'episode'])
df_COMMON_NAMES
```

```python
# Merging the two DataFrames
GCS_WITH_NAMES = pd.merge(df_GSC, df_COMMON_NAMES, on='episode')
GCS_WITH_NAMES.to_excel("Excel files/GCS_WITH_NAMES.xlsx", index=False)
GCS_WITH_NAMES
```

```python
Glasgow_values_whole_line = GCS_WITH_NAMES['GCS1'].values
pattern = r"(\d+)[/=](\d+)"

Glasgow_first_digit = []
line =2
unmatched = []
for text in Glasgow_values_whole_line:
    match = re.search(pattern, text)
    if match:
        Glasgow_first_digit.append( match.groups()[0])
    else:
        unmatched.append((line,text))
    line +=1

print(len(unmatched))
unmatched

```

```python
with open('unmatched.txt', 'w') as f:  # Changed 'unmatched.text' to 'unmatched.txt'
    for line_number, line_text in unmatched:
        f.write(f"Line {line_number}: {line_text}\n")
```

```python
import Levenshtein
import os

def find_similar_strings(target, string_list, threshold):
    similar_strings = []
    for string in string_list:
        distance = Levenshtein.distance(target, string)
        similarity_ratio = 1 - (distance / max(len(target), len(string)))
        if similarity_ratio >= threshold:
            similar_strings.append((string, similarity_ratio))
    return similar_strings 


PATH_TO_PATIENTS = r'C:\Users\user1\Desktop\PAGNI_TEST'
Glasgow_PATIENTS = GCS_WITH_NAMES_FINAL['NAMES_engl'].values
CORRESPOND_PATIENTS = []
counter =0
for patient in os.listdir(PATH_TO_PATIENTS):
    name = patient.split("_")[0]
    similar = find_similar_strings(name,Glasgow_PATIENTS,0.7)
    if similar!=[]:
        print("in PAGNI_TEST -->>",name ,"in GCS_WITH_NAMES_FINAL -->>",similar[0][0],"GCS :",  int( GCS_WITH_NAMES_FINAL[GCS_WITH_NAMES_FINAL['NAMES_engl']==similar[0][0]]['Glasgow_first_digit'].values[0]) )
        counter +=1

print(counter)
```
